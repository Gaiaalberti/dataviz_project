```python
import requests

def get_gender(name):
    url = f"https://api.genderize.io?name={name}"
    response = requests.get(url).json()
    if response["gender"]:
        return response["gender"]
    else:
        return None

list = []
for element in df.fullName:
    gender = get_gender(element.split()[0])
    list.append(gender)

```
